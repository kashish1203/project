```
%%file empsum.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, sum(salaries)

if __name__ == '__main__':
     EmpSum.run()
```

```
%%file empsum.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      gender=record[5]
      group=(dno,gender)
      yield group,1

    def reducer(self, group,count):
        yield group, sum(count)

if __name__ == '__main__':
     EmpSum.run()
```

```
%%file image_counter.py
from mrjob.job import MRJob

class ImageCounter(MRJob):
  def mapper(self,key,line):
    request = line.split(' ')
    ts=request[3]
    dsize=request[9]
    try:
      dsize=float(int(dsize)/pow(10,6))
    except:
      dsize=0
    finally:
      yield (ts[4:7],ts[8:12]),dsize

  def reducer(self,ts,size):
    yield ts,sum(size)

if __name__ == '__main__':
     ImageCounter.run()
```
